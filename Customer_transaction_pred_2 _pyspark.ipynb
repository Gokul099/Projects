```python
#import Randomforestclassifier
from pyspark.ml.classification import RandomForestClassifier
r = RandomForestClassifier(labelCol="target",featuresCol="features")
```

```python
split = df1.randomSplit([0.7,0.3])
```

```python

split[0].show(5)
```

```python
split[1].show()
```

```python
train = split[0]
test = split[1]
```

```python
train
```

```python
from pyspark.ml import Pipeline
```

```python
pipeline = Pipeline(stages=[va, r])
```

```python
model = r.fit(train)
```

```python
predictions = model.transform(test)
```

```python
predictions.show()
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

binary_evaluator = BinaryClassificationEvaluator(labelCol="target", rawPredictionCol="rawPrediction", metricName="areaUnderROC")
binary_auc = binary_evaluator.evaluate(predictions)
```

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
param_grid = ParamGridBuilder() \
    .addGrid(r.numTrees, [10, 20, 30]) \
    .addGrid(r.maxDepth, [5, 10, 15]) \
    .build()
```

```python
cross_validator = CrossValidator(estimator=pipeline,
                                 estimatorParamMaps=param_grid,
                                 evaluator=binary_evaluator,
                                 numFolds=3)
```

```python
eval_result = binary_evaluator.evaluate(predictions)
print(f"Area under ROC (Tuned Model): {eval_result}")
```

```python
from pyspark.ml.classification import LinearSVC
```

```python
svm = LinearSVC(featuresCol="features", labelCol="target", maxIter=10, regParam=0.01)
```

```python
predictions = svm_model.transform(test)

# Evaluate the model
evaluator = BinaryClassificationEvaluator(labelCol="target", rawPredictionCol="rawPrediction", metricName="areaUnderROC")
roc_auc = evaluator.evaluate(predictions)
print(f"Area under ROC: {roc_auc}")
```

```python
accuracy = evaluator.evaluate(predictions)

print(f"Accuracy: {accuracy}")
```
